```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
# Importing the Spanish model

model_name_es = "Helsinki-NLP/opus-mt-es-en"
model_es      = MarianMTModel.from_pretrained(model_name_es)
tokenizer_es  = MarianTokenizer.from_pretrained(model_name_es)
```

```python
# We compute the number of parameters

model_es.num_parameters()
```

```python
#Importing the French model

model_name_fr = "Helsinki-NLP/opus-mt-fr-en"
model_fr      = MarianMTModel.from_pretrained(model_name_fr)
tokenizer_fr  = MarianTokenizer.from_pretrained(model_name_fr)
```

```python
# The number of parameters in this case is:

model_fr.num_parameters()
```

```python
# Importing the German model

model_name_de = "Helsinki-NLP/opus-mt-de-en"
model_de      = MarianMTModel.from_pretrained(model_name_de)
tokenizer_de  = MarianTokenizer.from_pretrained(model_name_de)
```

```python
# The number of parameters will be:

model_de.num_parameters()
```

```python
# We can define a dictionary with the models and tokenizers. Could be done inside the function, I am not sure what's best


models = {
    'F': model_fr,
    'S': model_es,
    'G': model_de
}

tokenizers = {
    'F': tokenizer_fr,
    'S': tokenizer_es,
    'G': tokenizer_de
}
```

```python
# And we can define a function to translate


def translate(input_text, language):
  """
  Arguments:

  input_text: Text to be translated, a string
  language:   Input language: French ('F'), Spanish ('S') or German ('G')

  output: Translated text
  """
  model     = models[language]
  tokenizer = tokenizers[language]

  # The input text is encoded into tokens
  input_ids      = tokenizer.encode(input_text, return_tensors = "pt")

  # The tokens are then passed through the model
  translated_ids = model.generate(input_ids = input_ids, num_beams = 5, num_return_sequences = 1)

  # The tokens generated by the model are decoded to give the translated text
  output         = tokenizer.decode(translated_ids[0], skip_special_tokens = True, clean_up_tokenization_spaces = True)

  return output
```

Here are some examples of the function at work

```python
sentences_fr = ["Nous n'aimons pas le chocolat", "Je pense que non"]

for sentence in sentences_fr:
  print(f"French: {sentence}, French: {translate(sentence, 'F')}")
```

```python
sentences_es = ["Me gustan los caballos", "El examen de ayer era pan comido"]


for sentence in sentences_es:
  print(f"Spanish: {sentence}, English: {translate(sentence, 'S')}")
```

```python
sentences_ge = ["Ich esse Kekse", "Ich befolge die Regeln"]

for sentence in sentences_ge:
  print(f"German: {sentence}, English: {translate(sentence, 'G')}")
```
